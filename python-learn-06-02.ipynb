#### Numpy应用-2

```python
import matplotlib.pyplot as plt
import numpy as np
import random

array1 = np.random.randint(1,100,10)
array1
```

```python
#计算总和 均值 中位数
print(array1.sum())
print(np.sum(array1))
print(array1.mean())
print(np.mean(array1))
print(np.median(array1))
print(np.quantile(array1,0.5))#50百分位数

#方差 标准差 变异系数
print(array1.var())
print(np.var(array1))
print(array1.std())
print(np.std(array1))
print(array1.std()/array1.mean())
```

```python
#绘制箱型图
plt.boxplot(array1,showmeans=True)
plt.ylim([-20,120])
plt.show()
```

#### Numpy应用-3


```python
#numpy数组的运算会作用于每个元素
array1 = np.arange(1,10)
print(array1 + 10)
print(array1 * 10)
```

```python
#数组与数组运算
array2 = np.array([1,1,1,2,2,2,3,3,3])
print(array1+array2)
print(array1*array2)
print(array1**array2)
```

```python
print(array1>array2)
print(array1%array2==0)
```

```python
#通用一元函数
print(np.sqrt(array1))
print(np.log2(array1))
```

#### 广播机制


```python
#当对两个shape属性不同的数组进行加法
array5 = np.array([[0,0,0],[1,1,1],[2,2,2],[3,3,3]])
array6 = np.array([1,2,3])
array5 + array6
```

```python
array7 = np.array([[1],[2],[3],[4]])
array5 + array7
```

只有两个数组后缘维度相同或者后缘维度不同但其中一个数组后缘维度为1时，广播机制才会被触发。
所谓后缘维度，指的是数组形状（shape属性）从后往前看对应的部分

```python
array8 = np.array([1,2,3])
array9 = np.array([[1],[2],[3]])
array8 * array9
```
