#### 列表

````python
#索引(切片)
items9 = ['apple', 'waxberry', 'pitaya', 'peach', 'watermelon']
print(items9[0])
items9[2] = 'durian'
print(items9)
print(items9[-4])

print(items9[1:3:1])
print(items9[::2])#缺省值分别等于0和列表大小
print(items9[-4:-2])   # 缺省最后一项1
#切片修改
items9[1:3] = ['x', 'o']
print(items9)
````

````python
#列表的遍历
languages = ['Python', 'Java', 'C++', 'Kotlin']
for index in range(len(languages)):#len()返回列表的长度
    print(languages[index])
````

````python
#掷骰子次数统计
import random

counters = [0] * 6
# 模拟掷色子记录每种点数出现的次数
for _ in range(6000):
    face = random.randrange(1, 7)
    counters[face - 1] += 1
# 输出每种点数出现的次数
for face in range(1, 7):
    print(f'{face}点出现了{counters[face - 1]}次')
````

#### 列表的方法

````python
#append()入栈
languages = ['Python', 'Java', 'C++']
languages.append('JavaScript')
print(languages)
#insert()插入
languages.insert(1,'SQL')
print(languages)
#remove()移除
languages.remove('Java')
print(languages)
#弹出
temp = languages.pop()
print(temp)
temp = languages.pop(1)
print(temp)
#清空
languages.clear()
print(languages)
````

````python
#index方法获得元素在列表中的索引位置
items = ['Python', 'Java', 'Java', 'C++', 'Kotlin', 'Python']
print(items.index('Python'))
# 从索引位置1开始查找'Python'
print(items.index('Python',1))
# 数一共有多少个元素
print(items.count('Python'))
````

````python
items = ['Python', 'Java', 'C++', 'Kotlin', 'Swift']
#sort()方法实现排序
items.sort()
print(items)
#倒序
items.reverse()
print(items)
````

#### 列表生成式

````python
items = []
for i in range(1,100):
    if i %3 == 0 or i%5 == 0:
        items.append(i)
print(items)
#等效表达
items1 = [i for i in range(1,100) if i % 3 == 0 or i % 5 == 0]
print(items1)
````

````python
nums1 = [35, 12, 97, 64, 55]
nums2 = []
for num in nums1:
    nums2.append(num ** 2)
print(nums2)
#等效表达
nums2 = [num**2 for num in nums1]
print(nums2)
````

#### 嵌套列表

````python
scores = [[95, 83, 92], [80, 75, 82], [92, 97, 90], [80, 78, 69], [65, 66, 89]]
print(scores[0])
print(scores[0][1])
````

````python
#创建嵌套列表(矩阵
scores = []
for _ in range(5):
    temp = []
    for _ in range(3):
        score = int(input('请输入: '))
        temp.append(score)
    scores.append(temp)
print(scores)
````

````python
import random

scores = [[random.randrange(60, 101) for _ in range(3)] for _ in range(5)]
print(scores)
#rand.sample()方法随机取样
print(smaple1:=random.sample(scores,1))
print(random.sample(smaple1[0],1))
````

#### 打包解包操作
对元组进行解包可以得到元组的各个元素

````python
a = 1,10,100
print(type(a))
print(a)
#解包
i,j,k = a
print(i,j,k)
i,j,*k = a
print(i,j,k)
#解包元素不对应
#i,j = a

#使用星号表达式正确解包
a = 1,10,100,1000
i,j,*k = a #将剩下的所有元素放到k列表中，只能出现一次*
print(i,j,k)
i,j,k,l,*m = a
print(i,j,k,l,m)
````
